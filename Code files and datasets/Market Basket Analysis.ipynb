```python
import pandas as pd
import numpy as np

# importing apriori and association rules:
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
groceries = pd.read_csv('Market Baset Analysis- Groceries Data set.csv')
```

Getting the list of all unique items from all trxns: 

```python
items = np.array(pd.unique(groceries[groceries.columns[1:]].values.ravel(order = 'K')))
items
```

### Creatine new_groceries dataframe for creating structured data supporting for association rules:

```python
new_groceries = pd.DataFrame()
new_groceries['Transaction_id'] = groceries['Transaction Id'] 
for i in items:
    new_groceries[i] = np.nan
```

```python
new_groceries.drop_duplicates(subset= 'Transaction_id', inplace = True)
```

```python


for i in new_groceries['Transaction_id']:
    
    for j in groceries.columns[1:]:
        
        if len(groceries.loc[groceries['Transaction Id'] == i, j].values) == 2:
            x1 = groceries.loc[groceries['Transaction Id'] == i, j].values[0]
            x2 = groceries.loc[groceries['Transaction Id'] == i, j].values[1]
            if x1 is not np.nan:
                new_groceries.loc[new_groceries['Transaction_id'] == i, x1] = 1
            if x2 is not np.nan:
                new_groceries.loc[new_groceries['Transaction_id'] == i, x2] = 1
        
        else:
            x = groceries.loc[groceries['Transaction Id'] == i, j].values[0]
            if x is not np.nan:
                new_groceries.loc[new_groceries['Transaction_id'] == i, x] = 1
        
```

```python
new_groceries.fillna(0, inplace=True)
```

```python
new_groceries.head(7)
```

```python
new_groceries.shape
```

### Generating frequent itemsets and rules:

```python
freq_itemsets = apriori(new_groceries.iloc[:,1:], min_support=0.01, use_colnames=True)
```

```python
freq_itemsets.head(7)
```

```python
rules = association_rules(freq_itemsets, metric="lift", min_threshold=1)
```

```python
rules.head()
```

```python
rules.shape
```

```python
rules.sort_values(['lift','confidence'], ascending=False)
```

```python
# Filtering using the lift of greater than or equal to 2 and 
# confidence scores of greater than equal to 0.3:
rules[(rules['lift'] >= 2) & (rules['confidence'] >= 0.2)].sort_values(['lift','confidence'], ascending=False)
```
